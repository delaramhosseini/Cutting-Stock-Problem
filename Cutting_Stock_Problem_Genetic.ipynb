implementation of cutting stock problem with genetic algorithm.

1) At the first class is initialized with several parameters: maximum_itrations, num_of_population, and randomness. These parameters determine the maximum number of iterations for the algorithm, the number of individuals in the population, and the probability of mutation, respectively.


2) The input function takes a file path as input and reads the data from the file. The data includes the length of a rolls, a list of requests, and the expected number of rolls. This data is stored in instance variables for later use.


3) The creat_population function generates an initial population of chromosomes. Each chromosome is a random permutation of the indices representing the requests.

4) The fitness_function function calculates the fitness of a each chromosome. In this case, the fitness is the number of rolls required to fulfill all the requests in the chromosome. The method iterates over the requests in the chromosome, keeping track of the current sum of pieces. If the sum exceeds the length of a roll, a new roll is used.

5) The initial_fitnesse_of_each_chromosome function calculates the fitness for each chromosome in the initial population and returns a list of fitness values.

6) The min_fitness function finds the chromosome with the minimum fitness value in the current population and returns both the fitness value and the index of the chromosome.

7) The creat_new_generation function generates a new generation of chromosomes through a process of selection, crossover, and mutation. It randomly selects a subset of chromosomes from the current population, finds the best chromosomes within that subset, and uses them as parents to create a new child chromosome. The new child chromosome undergoes mutation with a certain probability. This process is repeated until the desired number of new chromosomes is generated.

8) The creat_new_child function performs crossover between two parent chromosomes to create a new child chromosome. It randomly selects a section of the parents' chromosomes and copies that section to the child chromosome. The remaining positions in the child chromosome are filled with the remaining requests in the order they appear in the other parent chromosome.

9) The creat_mutation function applies mutation to a chromosome. It randomly selects two positions in the chromosome and swaps the values at those positions with a certain probability.

10) At the end the run function is the main loop of the genetic algorithm. It iteratively creates new generations until either the maximum number of iterations is reached or the fitness of the best chromosome improves. In each iteration, it creates a new generation, evaluates the fitness of the new generation, and updates the current best solution if a better solution is found.


```python
import random
from copy import deepcopy 
```

```python
class genetic:
    def __init__(self , maximum_itrations = 1000, num_of_population = 30, randomness=0.8):
        self.randomness = randomness
        self.num_of_population = num_of_population
        self.number_of_requests = 0
        self.requests = []
        self.length_of_rolls = 0
        self.current_answer = []
        self.current_fitness = 0
        self.population = []
        self.fitness_of_each_chromosome = []
        self.expected_number_of_rolls = 0
        self.maximum_itrations = maximum_itrations
        
        
    def input(self , path):
        file = open(path, 'r')
        lines = file.readlines()
        length_of_roll  = int(lines[0].split()[-1])
        requests = list(map(int,lines[3][:-1].split(', ')))
        expected_number_of_rolls =  int(lines[5].split()[-1])
        self.requests = requests
        self.length_of_rolls = length_of_roll
        self.expected_number_of_rolls = expected_number_of_rolls
        self.number_of_requests = len(requests)
        self.population = self.creat_population()
        self.fitness_of_each_chromosome = self.initial_fitnesse_of_each_chromosome()
        min_fitness =  self.min_fitness()
        self.current_fitness = min_fitness[0]
        self.current_answer = self.population[min_fitness[1]]
        
       
    def creat_population(self):
        population = []
        for _ in range(self.num_of_population):
            chromosome = random.sample(range(self.number_of_requests),self.number_of_requests)
            population.append(chromosome)
            
        return population
    
    
    def fitness_function(self , chromosome): 
        requests = self.requests
        used_rolls = 1
        current_sum_of_pieces = 0
        for index in chromosome:
            current_sum_of_pieces+= requests[index]
            if (current_sum_of_pieces > self.length_of_rolls):
                used_rolls +=1
                current_sum_of_pieces = requests[index]
            
        return used_rolls 


    def initial_fitnesse_of_each_chromosome(self):
        fitness_of_each_chromosome = []
        for chromosome in self.population:
            fitness_of_each_chromosome.append(self.fitness_function(chromosome))
            
        return fitness_of_each_chromosome
    
    
    def min_fitness(self):
        min_fitness = self.fitness_of_each_chromosome[0]
        min_index = 0
        for i in range(1,self.num_of_population):
            if( min_fitness > self.fitness_of_each_chromosome[i]):
                min_fitness = self.fitness_of_each_chromosome[i]
                min_index = i
        
        return( min_fitness , min_index)
        
        
    # In this step we want to creat a new generation by using two parents, 
    # First, we make a random list as a index of population matrix that let us choose some random chromosomes from population,
    # Second, we will find the best fitness of those chromosomes that we randomly choosed,
    # Third, we choose the best chromosome as a parent, and we do this third steps for creating the second parent,
    # Fourth, we use creat_new_child function to great a child from these two parents.
    # Fifth, after creating a new child we should creat a mutation on this child for making a new generation, we do it with creat_mutation function.
    # sixth, after creating a new generation, indeed, making new chromosomes as equal to the length of population, we should find the fitness of this generation.
    # At the end, compare each chromosome in new generation to their parents and maintain the one that has a better fitness.
    def creat_new_generation(self, random_chromosome_number=3):
        new_generation = []
        iterate = 0
        while(iterate < self.num_of_population):
                
            random_chromosome_list1 = random.sample(range(self.num_of_population),random_chromosome_number)   
            best_chromosome1 = random_chromosome_list1[0]
            parent1 = []
            for chromosome_index in range(1,random_chromosome_number):
                if (self.fitness_of_each_chromosome[chromosome_index] <= self.fitness_of_each_chromosome[best_chromosome1]): 
                    best_chromosome1 = chromosome_index
                    
            parent1 = self.population[best_chromosome1]
            
            
            random_chromosome_list2 = random.sample(range(self.num_of_population),random_chromosome_number)
            best_chromosome2 = random_chromosome_list2[0]
            parent2 = []
            for chromosome_index in range(1,random_chromosome_number):
                if (self.fitness_of_each_chromosome[chromosome_index] <= self.fitness_of_each_chromosome[best_chromosome2]): 
                    best_chromosome1 = chromosome_index
                    
            parent2 = self.population[best_chromosome2]
            
            new_child = self.creat_new_child(parent1, parent2)
            
            new_generation.append(self.creat_mutation(new_child,))
            
            iterate += 1
        
        new_generation_fitness = []
        for i in range(self.num_of_population):
            new_generation_fitness.append(self.fitness_function(new_generation[i]))
            
        for i in range(self.num_of_population):
            if (new_generation_fitness[i] < self.fitness_of_each_chromosome[i]):
                self.population[i] = new_generation[i]
         
        self.fitness_of_each_chromosome = self.initial_fitnesse_of_each_chromosome()
        
        
        
    
    # For creating a child we use both parents together. In real, we use two chromosomes togheter.
    # At first, we creat tow children from tow chromosomes, then we choose the one which has a better fitness.
    def creat_new_child(self, parent1, parent2):
        length_of_parents = self.number_of_requests
        rand1 = random.randint(0, (length_of_parents - 2))
        rand2 = random.randint((rand1 + 1), (length_of_parents - 1))
        child1 = [ -1 for i in range(length_of_parents) ]
        child2 = [ -1 for i in range(length_of_parents) ]
        
        for i in range(rand1, rand2+1):
            child1[i] = parent1[i]
        
        for i in range(rand1, rand2+1):
            child2[i] = parent2[i]
        

        for i in range(length_of_parents):
            if child1[i] == -1:
                if parent2[i] not in child1:
                    child1[i] = parent2[i]

            if child2[i] == -1:
                if parent1[i] not in child2:
                    child2[i] = parent1[i]        

        for i in range(length_of_parents):
            if child1[i] == -1:
                for j in range(length_of_parents):
                    if j not in child1:
                        child1[i] = j
            if child2[i] == -1:
                for j in range(length_of_parents):
                    if j not in child2: 
                        child2[i] = j
                
        return child1 if (self.fitness_function(child1) <  self.fitness_function(child2)) else child2
            
            
    
    
    def creat_mutation(self , child):
        if random.random()<self.randomness:
            mutation_number = random.randint(0,int(0.2 * self.length_of_rolls))
            for _ in range (mutation_number):
                temp1 = random.randint(0, self.number_of_requests - 1)
                temp2 = random.randint(0, self.number_of_requests - 1)
                child[temp1],child[temp2] = child[temp2],child[temp1] 
                
        return child
        
          
    def run(self):
        itration = 0
        while (itration < self.maximum_itrations and self.current_fitness > self.expected_number_of_rolls):
            itration +=1
            self.creat_new_generation(3)
            min_fitness = self.min_fitness()
            fitness = min_fitness[0]
            answer = self.population[min_fitness[1]]
            
            if (fitness > self.current_fitness):break
            else : 
                if (fitness < self.current_fitness): print (f"in itration number {itration} best fitness is {fitness} ")
                self.current_fitness = fitness 
                self.current_answer = answer
                
    def print_answer(self): 
        print("final answer is : ")
        requests = self.requests
        used_rolls = 1
        current_sum_of_pieces = 0
        current_cutting = []
        for index in self.current_answer:
            current_sum_of_pieces+= requests[index]
            current_cutting.append(index)
            if (current_sum_of_pieces > self.length_of_rolls):
                print(f"{used_rolls} : {current_cutting[:-1]}")
                used_rolls +=1
                current_sum_of_pieces = requests[index]
                current_cutting=current_cutting[-1:]
        print(f"{used_rolls} : {current_cutting[:]}")
        
        
    def prnt(self):
        print (f"number_of_requests is : {self.number_of_requests}")
        print (f"requests is : {self.requests}")
        print (f"length_of_roll is : {self.length_of_rolls }")
        print (f"current_answer is : {self.current_answer}")
        print (f"current_fitness is : {self.current_fitness }")
        print (f"expected_nember_of_rolls is : {self.expected_number_of_rolls }")

        
```
